```python
from music21 import *
# for listing down the file names
import os
from fractions import *

# Array Processing
import numpy as np
```

```python

# defining function to read MIDI files
def read_midi(file):
    print("Loading Music File:", file)
    
    unparsed_notes= []


    # parsing a midi file
    midi_data = converter.parse(file)
    
    for part in midi_data.parts:
        print(part.partName)
        
        if part.partName == 'MELODY':
            notes_to_parse = part.recurse()
            
            unparsed_notes.append(notes_to_parse)
            
            
            


    return np.array(unparsed_notes)
```

```python

# specify the path
path = 'POP/'

files = []
limit = 1000


for index, i in enumerate(os.listdir(path)):
    if index == limit:
        break
    for j in os.listdir(path+i):
        file_path = (i+'/'+j)
        if file_path.endswith(".mid"):
            files.append(file_path)
        

print(files)

```

```python
# reading each midi file
unparsed_notes_array = []


for index, i in enumerate(files):
    if index == limit:
        break
    unparsed_notes_array.append(read_midi(path + i))
    

```

```python
unparsed_notes_array
```

```python

# processing unparsed notes


notes_array = []

for notes_to_parse in unparsed_notes_array:
    for element in notes_to_parse[0]:
        # note
        if isinstance(element, note.Note):
            d = duration.Duration()
            d.quarterLength = element.quarterLength
            notes_array.append([str(element.pitch),  f"{d.type}={element.quarterLength}"])


```

```python
notes_array
```

```python
notes_frequency = {}
duration_frequency = {}

for notes in notes_array:
    if notes[0] not in notes_frequency:
        notes_frequency[notes[0]] = 0
    notes_frequency[notes[0]]+=1
    if notes[1] not in duration_frequency:
        duration_frequency[notes[1]] = 0
    duration_frequency[notes[1]]+=1
        
```

```python
notes_frequency =  {key: val for key, val in sorted(notes_frequency.items(), key = lambda ele: -ele[1])}
notes_frequency
```

```python
duration_frequency =   {key: val for key, val in sorted(duration_frequency.items(), key = lambda ele: -ele[1])}
duration_frequency
```
